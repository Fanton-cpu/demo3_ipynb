# Importar los datos

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
flour = pd.read_csv('./data/fish_flour.csv')
```

# Exploración de datos

```python
flour.head()
```

```python
flour.info()
```

# Limpieza y preparación de datos

```python
# Asegurarse de que los tipos de datos sean adecuados
flour['Sample'] = flour['Sample'].astype(int)   # Convertir a tipo int si es necesario
```

# Análisis estadístico

```python
flour.describe()
```

# Análisis comparativo

```python
# Haremos una comparación con cada uno de los nutrientes de la harina de pescado
sns.set(style = 'whitegrid')
fig, axes = plt.subplots(2,2, figsize = (10,8))
plt.subplots_adjust(hspace = 0.5)
sns.barplot(ax = axes[0,0], data = flour , x = 'Sample' , y = 'Protein')
axes[0,0].set_title('Protein')
sns.barplot(ax = axes[0,1], data = flour , x = 'Sample' , y = 'Fat')
axes[0,1].set_title('Fat')
sns.barplot(ax = axes[1,0], data = flour , x = 'Sample' , y = 'Minerals')
axes[1,0].set_title('Minerals')
sns.barplot(ax = axes[1,1], data = flour , x = 'Sample' , y = 'Calcium')
axes[1,1].set_title('Calcium')
plt.show()
```

# Análisis de tendencias

```python
# Agregaremos fechas aleatorias en las que se sacaron las muestras
flour['Year'] = np.random.choice([2023,2022], size = len(flour))
flour['Month'] = np.random.choice(range(1,13) , size = len (flour))
flour['Day'] = np.random.choice(range(1,32) , size = len(flour))
flour['Date'] = pd.to_datetime(flour[['Year', 'Month', 'Day']].assign())
flour.sort_values('Date')
```

```python
# Ahora veremos la tendencia de costo en el mercado relacionando la meustra y el precio
sns.lineplot( data = flour , x = 'Sample' , y = 'Price')
plt.title('Cost trend in the market')
plt.xlabel('Sample')
plt.ylabel('Cost')
plt.show()
```
